```python
# Part 2_ Missing value stratgies for Non series
df1 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Question_6_part 2.csv')
df1

```

```python
df1.shape
```

```python
df1.head()
```

```python
df1.isnull().sum()
```

```python
#Applying mean strategy in dataset to fix null values
nt_case1 = df1
nt_case1['Age'] = nt_case1['Age'].replace(np.NaN,nt_case1['Age'].mean())
nt_case1
```

```python
# checking
nt_case1.isnull().sum()
```

```python
#Applying forward fill strategy in dataset to fix null values
nt_case2 = df1
nt_case2['Age'] = nt_case2['Age'].fillna(method='ffill')
nt_case2
```

```python
# checking
nt_case2.isnull().sum()
```
